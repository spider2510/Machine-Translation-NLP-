# Rearranging Tree

```python
def rearrange(tree1,level,res):
    if type(tree1) != list :
        if (len(tree1)==1):
            k=tree1[0]
            if type(k)!=nltk.tree.Tree:
                res.append(k)
    for subtree in tree1 :
        if type(subtree) == nltk.tree.Tree:
            if level>1:
                if (len(subtree)==2):
                    subtree[0],subtree[1]=subtree[1],subtree[0]
            rearrange(subtree,level+1,res)
```

```python
def inorder(tree,res):
    if (len(tree)==1):
        if (type(tree[0])!=nltk.tree.Tree):
            res.append(tree[0])
    for subtree in tree:
        if (type(subtree)==nltk.tree.Tree):
            if (subtree.label()=="VP"):
                post_order(subtree,res)
            else:
                inorder(subtree,res)
```

```python
def post_order(tree,res):
    stack=[]
    if (len(tree)==1):
        if (type(tree[0])!=nltk.tree.Tree):
            res.append(tree[0])
    for subtree in tree:
        stack.append(subtree)
    
    while (len (stack)!=0):
        k=stack.pop()
        if (type(k)==nltk.tree.Tree):
            if (k.label()=="VP"):
                post_order(k,res)
            else:
                inorder(k,res)
        
```

```python
def is_singular(en):
    p = inflect.engine()
    if p.singular_noun(en) is False:
        return(True)
    else:
        return(False)
```

```python
def determine_of(en):
    if (is_singular(en)):
        return "chyaa"
    else:
        return "che"
```

```python
def list_to_string(input_list):
    s=""
    for i in input_list:
        s=s+" "+i
    return(s)
```

```python
def translate_the_google(s):
    string=translator.translate(s,dest="mr",src="en").text 
    print(string)
    tokens = word_tokenize(string)
    return(tokens)
```

```python
def trans_marathi(res):
    s=""
    for i in range(len(res)):
        skip=0
        if (res[i]=="of" and i>0):
            res[i]= determine_of(res[i-1])
        if (res [i] in pronouns):
            for test in inflections_pronoun_detectors:
                if test in res:
                    s= s+ exceptional_pronoun[res[i]]
                    skip=1
        if (skip==0):
            s=s+" " +translator.translate(res[i],dest="mr",src="en").text
    return(s)
```

```python
def our_model(input_list):
    s1=list_to_string(input_list)
    tree = parser.raw_parse(s1)
    tree_list=list(tree)
    res=[]
    res1=[]
    rearrange(tree_list,0,res1)
    post_order(tree_list,res)
    remove_det(res)
    remove_det(res1)
    s1=trans_marathi(res)
    s2=trans_marathi(res1)
    tokens1 = word_tokenize(s1)
    tokens2 = word_tokenize(s2)
    
    return(tokens1,tokens2)
```
